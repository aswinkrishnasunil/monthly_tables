# Monthly data

```python
from tabula.io import read_pdf
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
```

```python

#creates a SQLAlchemy engine object that is connected to the SQLite database file db_monthly.db.
engine = create_engine('sqlite:///db_monthly.db')

#This line reads the PDF file forecast.pdf and extracts all of the tables from it into a list of Pandas DataFrames.
pdf_path = "forecast.pdf"
table = read_pdf(pdf_path, pages="all")

# Create a list to store all tables containing the given column
tables_with_given_column = []

# Iterate over the tables and add them to the list if they contain the given column
for i, table in enumerate(table):
    if all(column in table.columns for column in ['May-22', 'Jun-22', 'Jul-22', 'Aug-22', 'Sep-22', 'Oct-22', 'Nov-22', 'Dec-22', 'Jan-23', 'Feb-23', 'Mar-23', 'Apr-23']):
        tables_with_given_column.append(( table))


# Print all the tables into DataFrames
i=0
for  table in tables_with_given_column:
    name=table.columns  #The name variable is set to the list of column names in the current table.
    a=name[0]   #The a variable is set to the first column name in the name list.
    print(f"Table:{i+1}:") # prints the table name
    print(a)
    print(table)
    i=i+1
    
    #saves the current table to the SQLite database as a table with the name a. 
    #The con parameter specifies the SQLAlchemy engine object that is connected to the database.
    #The if_exists='replace' parameter ensures that the existing table is replaced with the new table.
    #The index=False parameter
    
    table.to_sql(name=a,con=engine,if_exists='replace', index=False)
```
